```python
import numpy as np
from scipy.io import loadmat  # this is the SciPy module that loads mat-files
import matplotlib.pyplot as plt
from datetime import datetime, date, time
from collections import OrderedDict
import pandas as pd
import numpy as np
import os
```

```python
directory="result"
subjects_dict=OrderedDict()
for filename in os.listdir(directory):
    mat = loadmat(os.path.join(directory, filename))  # load mat-file
    subject_resps=np.concatenate( mat['resp'], axis=0 )
    subject_times=np.concatenate( mat['time'], axis=0 )
    subject_inds=np.concatenate( mat['ind'], axis=0 )
    data={'resp':subject_resps, 'time':subject_times, 'index':subject_inds}
    subject_df = pd.DataFrame(data=data)
    subject_df = subject_df.sort_values('index').reset_index(drop=True)
    subjects_dict[filename.split('.')[0][-1]]=subject_df
```

```python
### Combine the test result of each testing subject into a joint array dataframe
joint_subjects_df = subjects_dict[subjects_dict.keys()[0]]
joint_resp_list=np.array(joint_subjects_df['resp'].tolist())
joint_resp_list=joint_resp_list.reshape(joint_resp_list.shape[0],-1)
joint_time_list=np.array(joint_subjects_df['time'].tolist())
joint_time_list=joint_time_list.reshape(joint_time_list.shape[0],-1)
for key, values in subjects_dict.items():
    ind_list=values['index'].tolist()
    if key != subjects_dict.keys()[0]:
        value_resp_list=np.array(values['resp'].tolist())
        value_resp_list=value_resp_list.reshape(value_resp_list.shape[0],-1)
        joint_resp_list=np.concatenate((joint_resp_list,value_resp_list), axis=1)
        value_time_list=np.array(values['time'].tolist())
        value_time_list=value_time_list.reshape(value_time_list.shape[0],-1)   
        joint_time_list=np.concatenate((joint_time_list,value_time_list), axis=1)

data={'resps':joint_resp_list.tolist(),'times':joint_time_list.tolist(),'index':ind_list}
joint_subject_df = pd.DataFrame(data=data)

joint_subject_df['mean_time']=np.mean(np.array(joint_subject_df['times'].tolist()),axis=1).tolist()
joint_subject_df.head(10)
```
